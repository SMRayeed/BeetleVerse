```python
import os
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
import warnings
warnings.simplefilter("ignore")
```

```python
file_path = "/fs/ess/PAS2136/Hawaii-2025/beetles_intake/BeetlePalooza Data/CSV Update/individuals_segmented_run1_updated.csv"
df = pd.read_csv(file_path)
df = df.drop(columns=['scientificName'])
df['scientificName'] = df['genus'].str.split().str[0].str.capitalize() + " " + df['species'].str.split().str[0].str.lower()
df.head(2)
```

```python
len(df)
```

```python
df = df.dropna(subset=['individual_image_file_path']).drop_duplicates(subset=['individual_image_file_path'])
len(df)
```

```python
cols = [
    'individual_image_file_path',
    'beetleID',
    'pictureID',
    "length_distance_cm",
    "width_distance_cm",
    "scientificName",
    "genus",
    "species",
    'siteID',
    'collection_date',
    'plotID',
    'siteName',
    'siteState',
    'siteLatitude',
    'siteLongitude',
    'siteType',
    'siteTerrain',
    'evapotranspiration',
    'max_temp',
    'min_temp',
    'precipitation',
    'precipitation_hours',
    'rain_sum',
    'shortwave_radiation',
    'siteDomainCode',
    'snowfall_sum',
    'sunshine_duration',
    'weather_code',
    'wind_gusts_max',
    'wind_speed_max'
]
new_df = df[cols]
```

```python
new_df.head(2)
```

```python
new_df.loc[:, "individual_image_file_path"] = "/fs/ess/PAS2136/Rayeed/BeetlePalooza/" + new_df["individual_image_file_path"]
new_df.head(3)
```

```python
output_dir = "/fs/ess/PAS2136/Hawaii-2025/beetles_intake/BeetlePalooza Data/Benchmarking"
os.makedirs(output_dir, exist_ok=True)
new_df.to_csv(os.path.join(output_dir, "Beetlepalooza_beetles.csv"), sep="\t", index=False)
```

```python
df_images_only = new_df[["individual_image_file_path", "genus", "scientificName"]]
df_images_only.rename(columns={
    "individual_image_file_path": "ImageFilePath",
    "genus": "Genus",
    "scientificName": "ScientificName"
}, inplace=True)
df_images_only.head(2)
```

```python
df_images_only.to_csv(os.path.join(output_dir, "Beetlepalooza_beetles_image_only.csv"), sep="\t", index=False)
```

```python
df_images_measurement = new_df[["individual_image_file_path", "genus", "scientificName", "length_distance_cm", "width_distance_cm"]]
df_images_measurement.rename(columns={
    "individual_image_file_path": "ImageFilePath",
    "genus": "Genus",
    "scientificName": "ScientificName",
    "length_distance_cm" : "Length",
    "width_distance_cm" : "Width"
}, inplace=True)
df_images_measurement.head(2)
```

```python
df_images_measurement.to_csv(os.path.join(output_dir, "Beetlepalooza_beetles_image_measurements.csv"), sep="\t", index=False)
```

```python
df_images_weather = new_df[[
    "individual_image_file_path",
    "genus",
    "scientificName",                        
    'siteLatitude',
    'siteLongitude',
    'evapotranspiration',
    'max_temp',
    'min_temp',
    'precipitation',
    'precipitation_hours',
    'rain_sum',
    'shortwave_radiation',
    'siteDomainCode',
    'snowfall_sum',
    'sunshine_duration',
    'weather_code',
    'wind_gusts_max',
    'wind_speed_max']]

df_images_weather.rename(columns={
    "individual_image_file_path": "ImageFilePath",
    "genus": "Genus",
    "scientificName": "ScientificName"
}, inplace=True)

df_images_weather.head(2)
```

```python
df_images_weather.to_csv(os.path.join(output_dir, "Beetlepalooza_beetles_image_df_weather.csv"), sep="\t", index=False)
```
